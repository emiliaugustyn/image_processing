### Wypróbuj działanie wyrównywania histogramu na przykładowych obrazach. By zaobserwować skuteczność procedury, poddaj wyrównywaniu obrazyzbyt ciemne i zbyt jasne. Narysować histogramy obrazów przed i po wyrównaniu.

```python
import numpy as np 
import cv2 as cv
import matplotlib.pyplot as plt

img_original = cv.imread("balloons.jpg",0)

def show_images(first, second):
    fig, axes = plt.subplots(1, 2, figsize=(20, 20))
    ax = axes.ravel()
    ax[0].imshow(first, cmap=plt.cm.gray)
    ax[1].imshow(second, cmap=plt.cm.gray)
```

### 1. Wyświetlenie obrazu zbyt ciemnego (lewa strona) i zbyt jasnego (prawa strona):

```python
gamma_light = 0.5
gamma_dark = 10
img_light = np.array(255*(img_original / 255) ** gamma_light, dtype = 'uint8')
img_dark = np.array(255*(img_original / 255) ** gamma_dark, dtype = 'uint8')
show_images(img_dark,img_light)
```

### 2. Wyświetlenie histogramów obrazów, ciemnego po lewej, jasnego po prawej, stronie:


```python
fig, axes = plt.subplots(1, 2, figsize=(20, 5))
axes[0].hist(img_dark.ravel(), 256)
axes[1].hist(img_light.ravel(), 256)
plt.show()
```
