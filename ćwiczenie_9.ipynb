### Zastosuj filtry górnoprzepustowe do wykrywania krawędzi na przykładowych obrazach.

```python
import numpy as np 
import cv2 as cv
import matplotlib.pyplot as plt

img_original = cv.imread("balloons.jpg",0)

def show_images(first, second):
    fig, axes = plt.subplots(1, 2, figsize=(20, 20))
    ax = axes.ravel()
    ax[0].imshow(first, cmap=plt.cm.gray)
    ax[1].imshow(second, cmap=plt.cm.gray)
```

### 1. Filtr z maską Sobela do wykrywania krawędzi, kolejno od lewej: oryginalny obraz, sobel x, sobel y:

```python
sobelx = cv.Sobel(img_original,cv.CV_64F,1,0,ksize=5)  
sobely = cv.Sobel(img_original,cv.CV_64F,0,1,ksize=5)  

fig, axes = plt.subplots(1,3,figsize=(20, 20))
ax = axes.ravel()
ax[0].imshow(img_original, cmap=plt.cm.gray)
ax[1].imshow(sobelx, cmap=plt.cm.gray)
ax[2].imshow(sobely, cmap=plt.cm.gray)
    
```

### 2. Działanie Laplasjanu w wykrywaniu krawędzi:

```python
laplacian = cv.Laplacian(img_original,cv.CV_64F)
show_images(img_original,laplacian)
```
