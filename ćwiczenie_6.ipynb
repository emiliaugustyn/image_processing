### Zaobserwuj działanie następujących przekształceń punktowych na przykładowych obrazach.


```
import numpy as np 
import cv2 as cv
import matplotlib.pyplot as plt

img_original = cv.imread("balloons.jpg",0)

def show_images(first, second):
    fig, axes = plt.subplots(1, 2, figsize=(20, 20))
    ax = axes.ravel()
    ax[0].imshow(first, cmap=plt.cm.gray)
    ax[1].imshow(second, cmap=plt.cm.gray)
```

### 1. Mnożenie obrazu przez stałą T(r)  = c·r, gdzie c jest stałą:

```
img_multiplied = cv.multiply(img_original,2)
show_images(img_original,img_multiplied)
```

### 2. Zmiana dynamiki skali szarości (kontrastu):

```
m = 0.45
e = 8
r = 1 / (1 + (m / img_original) ** e)
show_images(img_original, r)
```

### 3. Korekcja gamma:

```
gamma = 0.1
img_gammacorrection = np.array(255*(img_original / 255) ** gamma, dtype = 'uint8')
show_images(img_original,img_gammacorrection)
```
