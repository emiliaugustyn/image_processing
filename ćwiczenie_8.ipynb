### Badanie skuteczności redukcji szumu typu "sól i pieprz".

```python
import numpy as np 
import cv2 as cv
import matplotlib.pyplot as plt

img_original = cv.imread("tygrysek.jpg",0)

def show_images(first, second):
    fig, axes = plt.subplots(1, 2, figsize=(20, 20))
    ax = axes.ravel()
    ax[0].imshow(first, cmap=plt.cm.gray)
    ax[1].imshow(second, cmap=plt.cm.gray)
```

### 1. Liniowy filtr uśredniający z kwadratową maską, rozpoczynając od maski rozmiaru 3×3:


```python
kernel = np.ones((3,3),np.float32)/9
img_boxfilter = cv.filter2D(img_original,-1,kernel)
show_images(img_original,img_boxfilter)
```

### 2. Nieliniowy filtr medianowy:

```python
img_medianfilter = cv.medianBlur(img_original,5)
show_images(img_original,img_medianfilter)
```
